**Credit Status Analysis: Impact of Employment Duration**

```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np


# Read csv into a DataFrame
df_sample = pd.read_csv("/Users/pm300/Downloads/sample_dataset (1).csv")
```

```python
bins = [0,5,10,15,20,50]
labels = ['0-5', '5-10', '10-15', '15-20', '20+']
df_sample['Years Employed'] = pd.cut(df_sample['Years Employed'], bins=bins, labels=labels)
```

```python
cross_tab = pd.crosstab(df_sample['Years Employed'], df_sample['Credit Status'])
print(cross_tab)
```

```python


# Generate the stacked bar chart
ax = cross_tab.plot(kind='bar', stacked=True)

# Add value labels to the chart
ax.bar_label(ax.containers[0], label_type='center', color='white', weight='bold')
ax.bar_label(ax.containers[1], label_type='center', color='white', weight='bold')
ax.bar_label(ax.containers[2], label_type='center', color='white', weight='bold')
ax.bar_label(ax.containers[3], label_type='center', color='white', weight='bold')

# Set the x-axis label, y-axis label, and title
plt.xlabel('Employment Duration')
plt.ylabel('Count')
plt.title('Credit Status by Employment Duration')

# Rotate x-axis labels to avoid overlapping
plt.xticks(rotation=360)

# Display the chart
plt.show()
```

**Null Hypothesis (H0): There is no association between years employed and credit status**

**Alternative Hypothesis (Ha): There is a significant association between years employed and credit status**

```python
data = {'Credit Status': ['<30 days overdue', '>60 days overdue', 'Loan paid', 'No loans this month'],
'0-5': [8644, 429, 9062, 3980],
'5-10': [5364, 201, 6912, 2925],
'10-15': [2356, 92, 2741, 1403],
'15-20': [1095, 29, 1245, 467],
'20+': [1220, 47, 1245, 539]}
df_sample= pd.DataFrame(data)
df_sample.set_index('Credit Status', inplace=True)
chi2, p, dof, expected = stats.chi2_contingency(df_sample)
print("Chi-square statistic:", chi2)
print("p-value:", p)
print('if P-value <{0.05}, then accept the null')
```

**How Childeren affect credit card status**

```python

# Create a DataFrame with the provided data
data = {
    'Credit Status': ['<30 days overdue', '>60 days overdue', 'Loan paid', 'No loans this month'],
    'N': [11974, 503, 13597, 5869],
    'Y': [6708, 295, 7609, 3445]
}
df = pd.DataFrame(data)
df.set_index('Credit Status', inplace=True)

# Plot the grouped bar chart
ax = df.plot(kind='barh', stacked=True)

# Set labels and title
ax.set_xlabel('Children')
ax.set_ylabel('Count')
ax.set_title('Credit Status by Children')

# Add values on top of each bar
for container in ax.containers:
    ax.bar_label(container, label_type='center', fontsize=7, color ='white')

# Adjust the layout to fit the labels
plt.tight_layout()

# Show the plot
plt.show()
```

**Null Hypothesis (H0): There is no association between credit status and having children**

**Alternative Hypothesis (Ha): There is a significant association between credit status and having children**

```python
# Set up the contingency table
observed = np.array([[11974, 503, 13597, 5869], [6708, 295, 7609, 3445]])

# Perform chi-square test of independence
chi2, p, dof, expected = stats.chi2_contingency(observed)

print("Chi-square statistic:", chi2)
print("p-value:", p)
print(f'if P-value >{0.05}, then reject the null')
```
